# GENDER EXPLORATION

 I adapted the process used by Kaggler shivamnijhawan96: https://www.kaggle.com/shivamnijhawan96/mental-health-survey-exploration.

```python
df['Gender'] = df['Gender'].str.lower()
male = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "cis male"]
trans = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means"]
female = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]
df['Gender'] = df['Gender'].apply(lambda x:"Male" if x in male else x)
df['Gender'] = df['Gender'].apply(lambda x:"Female" if x in female else x)
df['Gender'] = df['Gender'].apply(lambda x:"Trans" if x in trans else x)
df.drop(df[df.Gender == 'p'].index, inplace=True)
df.drop(df[df.Gender == 'a little about you'].index, inplace=True)
```

## Chi-Squared Test of Independence
I would like to calculate whether these variables are statistically independent of each other. First, I will try a chi-squared test of independence. (Adapting the process from: http://hamelg.blogspot.com/2015/11/python-for-data-analysis-part-25-chi.html) 

```python
# Generate crosstabs of "family history" and "treatment" (the "observed" values)

import scipy.stats as stats

hipothisis = pd.crosstab(df["family_history"], df["treatment"], margins = True)
hipothisis.columns = ["Treatment Yes","Treatment No","row_totels"]
hipothisis.index = ["Fam Hist Yes","Fam Hist No","col_totels"]

observed = hipothisis.iloc[0:2,0:2]   # Get table without totals for later use

hipothisis
```

```python
# Generate the "expected" values to compare against our "observed" values

expected =  np.outer(hipothisis["row_totals"][0:2],
                     hipothisis.loc["col_totels"][0:2]) / 1257

expected = pd.DataFrame(expected)

expected.columns = ["Treatment Yes","Treatment No"]
expected.index = ["Fam Hist Yes","Fam Hist No"]

expected
```

```python
# Run the Chi-Squared test

chi_squared_stat = (((observed-expected)**2)/expected).sum().sum()
print(chi_squared_stat)
```

```python
crit = stats.chi2.ppf(q = 0.95, 
                      df = 1)   

# *Note: The degrees of freedom for a test of independence equals the product of 
# the number of categories in each variable minus 1. In this case we have a 2x2 table 
# so df = 1x1 = 1.

print("Critical value")
print(crit)

p_value = 1 - stats.chi2.cdf(x=chi_squared_stat,  # Find the p-value
                             df=1)
print("P value")
print(p_value)
```
